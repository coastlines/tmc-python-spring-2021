```python
import civis
import json
import pandas
import urllib.request
from pandas.io.json import json_normalize

URL = 'https://services.arcgis.com/su8ic9KbA7PYVxPS/arcgis/rest/services/New_Cases_Total_2/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json'

body = urllib.request.urlopen(URL).read()
parsed_body = json.loads(body)
df = pandas.json_normalize(parsed_body,record_path=['features'])
df.columns = df.columns.str.replace('attributes.', '')
```
