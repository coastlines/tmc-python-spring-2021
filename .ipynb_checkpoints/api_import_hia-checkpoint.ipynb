```python
import json
import pandas
import urllib.request
from pandas.io.json import json_normalize

URL = 'https://services.arcgis.com/su8ic9KbA7PYVxPS/arcgis/rest/services/HCPH_COVID19_Zip_Codes_download/FeatureServer/0/query?where=1%3D1&outFields=ZIP,TotalPop,TotalConfirmedCases,ActiveCases_str,Death_str,EditDate&returnGeometry=false&outSR=4326&f=json'
```

```python
body = urllib.request.urlopen(URL).read()
```

```python
parsed_body = json.loads(body)
```

```python
df = pandas.json_normalize(parsed_body,record_path=['features'])
```
